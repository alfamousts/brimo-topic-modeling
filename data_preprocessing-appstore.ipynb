```python
import pandas as pd
import os
import re
import nltk
import string
import emoji
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from collections import Counter
import dataPreprocessing
```

```python
df = pd.read_csv('./data/result-appstore-combine.csv',delimiter=',',encoding= 'unicode_escape')
```

```python
df
```

```python
stopWords = dataPreprocessing.createStopword()
```

```python
newContent = []
dataRawBOW = []
for index, row in df.iterrows():
    newText = dataPreprocessing.preProcessing(row['content'],stopWords)
    dataRawBOW.append(newText.split())
    newContent.append(newText)
```

```python
newContent
```

```python
df['newContent'] = newContent
```

```python
df.to_csv('./data/result-appstore-combine-preprocessed2.csv', index=False,encoding='utf-8')
```

```python
import csv
csvFile = open('./result/result_rawBOW_appstore.csv','w',newline='',errors='replace')
csvWriter = csv.writer(csvFile,delimiter=',',quotechar='"',quoting=csv.QUOTE_MINIMAL)
for word in dataRawBOW:
    if(len(word) > 0):
        csvWriter.writerow(word)
csvFile.close()
```
